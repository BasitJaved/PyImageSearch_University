```python
from google.colab.patches import cv2_imshow
from imutils.object_detection import non_max_suppression
import numpy as np
import cv2 as cv

img = 'image1.png'
tmplt = 'temp1.png'
thresh = 0.8

# load input image and templet and grab the template image spatial dimensions, and display images
print('[INFO] loading images...')
image = cv.imread(img)
template = cv.imread(tmplt)
(tH, tW) = template.shape[:2]
cv2_imshow(image)
cv2_imshow(template)

# convert both to greyscale
image_gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
template_gray = cv.cvtColor(template, cv.COLOR_BGR2GRAY)

# perform templet matching
print('[INFO] performing templet matching...')
result = cv.matchTemplate(image_gray, template_gray, cv.TM_CCOEFF_NORMED)

# find all locations in the result map where matched value is greater then threshold, then clone
# original image so we can draw on it
(yCoords, xCoords) = np.where(result>=thresh)
clone = image.copy()
print(f'[INFO] {len(yCoords)} matched locations before NMS')

# loop over the starting (x, y)-coordinates
for (x, y) in zip(xCoords, yCoords):
  # draw bounding boxes on image
  cv.rectangle(clone, (x, y), (x + tW, y + tH), (255, 0, 0), 3)

# show output image before applying NMS
cv2_imshow(clone)

# initialize our list of rectangles
rects = []

# loop over the starting (x, y)-coordinates again
for (x, y) in zip(xCoords, yCoords):
  # update our list of rectangles
  rects.append((x, y, x+tW, y+tH))

# apply non-maxima suppression to rectangles
pick = non_max_suppression(np.array(rects))
print(f'[INFO] {len(pick)} matched locations after NMS')

# loop over the starting (x, y)-coordinates
for (startX, startY, endX, endY) in pick:
  # draw bounding boxes on image
  cv.rectangle(image, (startX, startY), (endX, endY), (255, 0, 0), 3)

cv2_imshow(image)
```
