```python
from google.colab.patches import cv2_imshow
import cv2 as cv

img = 'image1.png'
tmplt = 'temp1.png'

# load input image and templet and display them
print('[INFO] loading images...')
image = cv.imread(img)
template = cv.imread(tmplt)
cv2_imshow(image)
cv2_imshow(template)

# convert both to greyscale
image_gray = cv.cvtColor(image, cv.COLOR_BGR2GRAY)
template_gray = cv.cvtColor(template, cv.COLOR_BGR2GRAY)

# perform templet matching
print('[INFO] performing templet matching...')
result = cv.matchTemplate(image_gray, template_gray, cv.TM_CCOEFF_NORMED)
(min_val, max_val, min_Loc, max_Loc) = cv.minMaxLoc(result)

# determin the staring and ending (x, y) coordinates of bounding box
(start_x, start_y) = max_Loc
end_x = start_x + template.shape[1]
end_y = start_y + template.shape[0]

# draw bounding box on image
cv.rectangle(image, (start_x, start_y), (end_x, end_y), (255, 0, 0), 3)

cv2_imshow(image)
```
